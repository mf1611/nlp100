```python
file_path = './output/neko.txt.mecab'

sentences = []
morphs = []
with open(file_path, 'rt') as f:
    for line in f:
        
        if line != 'EOS\n':  # 文末以外
            # 表層形(記載の語)が文頭で，\t区切りで，品詞・原形が　,区切りで続く
            fields = line.split('\t')
            if (fields[0]!='') & (fields[0]!='\n'):  # 空白文字と改行文字は対象外
                attr = fields[1].split(',')
                # 表層形, 原形，品詞，品詞細分類1
                morph = {'surface': fields[0], 'base': attr[6], 'pos': attr[0], 'pos1': attr[1]}
                morphs.append(morph)
                
        else:  # 文末で
            if len(morphs)!=0:  # EOSとEOSの間が空白で何もない場合がある
                sentences.append(morphs)
                morphs = []

sentences
```

```python
result = set()
for sentence in sentences:
    for i in range(1, len(sentence)-1):
        if (sentence[i-1]['pos']=='名詞') & (sentence[i]['surface']=='の') & (sentence[i+1]['pos']=='名詞'):
            result.add(sentence[i-1]['surface'] + sentence[i]['surface'] + sentence[i+1]['surface'])

result
```

```python
result = set()
for sentence in sentences:
    nouns = ''
    num = 0
    for morph in sentence:
        if morph['pos']=='名詞':
            nouns += morph['surface']
            num += 1
        elif num >= 2:  # 名詞なら上のifブロックだけ実行されこのブロックは実行されない
            result.add(nouns)
            nouns = ''
            num = 0
        else:
            nouns = ''
            num = 0
    
result
```

```python
word_count = dict()
for sentence in sentences:
    for morph in sentence:
        if morph['pos'] != '記号':
            if morph['base'] not in word_count.keys():
                word_count[morph['base']] = 1
            else:
                word_count[morph['base']] += 1

word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
word_count
```

- defaultdictを使う場合，上記のようなキーが存在しない場合の初期化をやる必要がない！

```python
from collections import defaultdict 

word_count = defaultdict(int)
for sentence in sentences:
    for morph in sentence:
        if morph['pos'] != '記号':
            word_count[morph['base']] += 1

word_count = dict(sorted(word_count.items(), key=lambda x: x[1], reverse=True))
word_count
```

```python
keyword = '猫'

result = defaultdict(int)
for sentence in sentences:
    if keyword in [morph['base'] for morph in sentence]:
        for morph in sentence:
            if (morph['pos']!='記号') & (morph['base']!=keyword):
                result[morph['base']] += 1

result = dict(sorted(result.items(), key=lambda x: x[1], reverse=True))
result
```
